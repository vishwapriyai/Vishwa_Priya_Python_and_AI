# K-means clustering

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.cluster import KMeans
%matplotlib inline
```

```python
# generating two cluster and plotting
x= -2*np.random.rand(100,2) # random.rand== get random number, 100 * 2 --- 100 array with 2 values
print(x)
x1=1+2*np.random.rand(50,2)
print(x1)
x[50:100,:]=x1
print(x)
plt.scatter(x[:,0],x[:,1],s=50,c='b')
plt.show()
# x[:,0],x[:,1]--- oth index and 1st index ---x,y
```

```python
Kmean=KMeans(n_clusters=2)
Kmean.fit(x)
```

```python
# centroid --- the centre of the data points of the specific clusters
Kmean.cluster_centers_
```

```python
plt.scatter(x[:,0],x[:,1],s=50,c='b')
plt.scatter(2.1171602 ,  2.13394506,s=200,c='g', marker='d')
plt.scatter(-1.02601352, -0.89138649,s=200,c='r', marker='s')
plt.show()

# s-size
#c-color
#markers='s'---- square
#markers='d'---- dimond
```

```python
Kmean.labels_
```

```python
# need to check where thsese data points in the cluster---3.0,3.0

sample_test=np.array([3.0,3.0])
second_test=sample_test.reshape(1,-1)
print(second_test)
Kmean.predict(second_test)
```

```python
sample_test=np.array([-3.0,3.0])
second_test=sample_test.reshape(1,-1)
print(second_test)
Kmean.predict(second_test)
```

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
%matplotlib inline
from sklearn.cluster import AgglomerativeClustering
```

```python
X=np.array([[5,3],
            [10,15],
            [15,12],
            [24,10],
            [30,30],
            [85,70],
            [71,80],
            [60,78],
            [70,55],
            [80,91],])
```

```python
from sklearn.cluster import AgglomerativeClustering
cluster=AgglomerativeClustering(n_clusters=2,metric='euclidean',linkage='ward')
cluster.fit_predict(X)
```

```python
print(cluster.labels_)
```

```python
plt.scatter(X[:,0],X[:,1], c=cluster.labels_,cmap='rainbow')
```
